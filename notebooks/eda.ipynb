```python
# IMPORTANT: The affordability_gap dataset has multiple rows per institution
# Each row represents a different "Student Family Earnings Ceiling" scenario
# This means columns like "Net Price" and "Affordability Gap" vary by earnings ceiling
# 
# When merging, we preserve ALL rows (not dropping duplicates) to maintain
# the granularity of institution-earnings ceiling combinations

merged = pd.merge(
    left=college_2021, 
    right=affordability_gap, 
    left_on="UNIQUE_IDENTIFICATION_NUMBER_OF_THE_INSTITUTION",
    right_on="Unit ID",
    how="inner"  # or "left" depending on analysis needs
)

print(f"College 2021 rows: {len(college_2021)}")
print(f"Affordability Gap rows: {len(affordability_gap)}")
print(f"Merged rows: {len(merged)}")
print(f"\nNote: Merged dataset has more rows because each institution from college_2021")
print(f"is duplicated for each Student Family Earnings Ceiling scenario in affordability_gap")

merged
```

```python
# Example: View data for a single institution across all earnings ceiling scenarios
example_institution = merged["UNIQUE_IDENTIFICATION_NUMBER_OF_THE_INSTITUTION"].iloc[0]
print(f"Institution ID: {example_institution}")
print(f"Institution Name: {merged[merged['UNIQUE_IDENTIFICATION_NUMBER_OF_THE_INSTITUTION'] == example_institution]['Institution Name'].iloc[0]}")
print("\nData across different Student Family Earnings Ceiling scenarios:")
merged[merged["UNIQUE_IDENTIFICATION_NUMBER_OF_THE_INSTITUTION"] == example_institution][[
    "UNIQUE_IDENTIFICATION_NUMBER_OF_THE_INSTITUTION",
    "Institution Name",
    "Student Family Earnings Ceiling",
    "Net Price",
    "Affordability Gap (net price minus income earned working 10 hrs at min wage)",
    "Cost of Attendance: In State, On Campus"
]]
```

```python
# EXAMPLE: How to get one row per institution by averaging across earnings ceilings
# This is useful when you want institution-level statistics

# Identify columns that vary by earnings ceiling (need aggregation)
earnings_dependent_cols = [
    "Net Price",
    "Affordability Gap (net price minus income earned working 10 hrs at min wage)",
    "Weekly Hours to Close Gap"
]

# Identify columns that are constant per institution (can take first value)
institution_constant_cols = [
    "Institution Name",
    "City", 
    "State Abbreviation",
    "Sector",
    "Control of Institution",
    "Cost of Attendance: In State, On Campus"
    # Add more as needed
]

# Create aggregation dictionary
agg_dict = {col: "mean" for col in earnings_dependent_cols}
agg_dict.update({col: "first" for col in institution_constant_cols})

# Aggregate to get one row per institution
merged_institution_level = merged.groupby("UNIQUE_IDENTIFICATION_NUMBER_OF_THE_INSTITUTION").agg(agg_dict).reset_index()

print(f"Original merged dataset: {len(merged)} rows")
print(f"Institution-level aggregated dataset: {len(merged_institution_level)} rows")
print(f"Number of unique institutions: {merged['UNIQUE_IDENTIFICATION_NUMBER_OF_THE_INSTITUTION'].nunique()}")

merged_institution_level.head()
```

```python
"""
WORKING WITH THE GRANULAR DATA STRUCTURE

Since each institution has multiple rows (one per Student Family Earnings Ceiling),
you have several options for analysis:

1. KEEP ALL ROWS: Analyze by earnings ceiling scenario
   - Use the full merged dataset as-is
   - Group by both institution AND earnings ceiling when aggregating
   
2. FILTER TO SPECIFIC EARNINGS CEILING: Focus on one income bracket
   - Example: merged[merged["Student Family Earnings Ceiling"] == 30000.0]
   - Valid values: 30000.0, 48000.0, 75000.0, 110000.0, 150000.0
   
3. AGGREGATE PER INSTITUTION: Get one row per institution
   - Example: Average across all earnings ceilings
   - merged.groupby("UNIQUE_IDENTIFICATION_NUMBER_OF_THE_INSTITUTION").agg({
         "Net Price": "mean",
         "Affordability Gap (net price minus income earned working 10 hrs at min wage)": "mean"
     })
"""

# Show the earnings ceiling categories available
print("Available Student Family Earnings Ceiling categories:")
print(merged["Student Family Earnings Ceiling"].value_counts().sort_index())
```

```python
# Examine the Student Family Earnings Ceiling values to understand granularity
print("Unique Student Family Earnings Ceiling values:")
print(affordability_gap["Student Family Earnings Ceiling"].unique())
print(f"\nNumber of rows per institution: {affordability_gap['Unit ID'].value_counts().mode()[0]}")
print(f"\nTotal institutions: {affordability_gap['Unit ID'].nunique()}")
print(f"Total rows: {len(affordability_gap)}")

# Show example of one institution with all its earnings ceiling scenarios
example_id = affordability_gap["Unit ID"].value_counts().index[0]
print(f"\n\nExample: Institution {example_id} has the following earnings ceiling scenarios:")
print(affordability_gap[affordability_gap["Unit ID"] == example_id][["Unit ID", "Institution Name", "Student Family Earnings Ceiling", "Net Price", "Affordability Gap (net price minus income earned working 10 hrs at min wage)"]])
```
